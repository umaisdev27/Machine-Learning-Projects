```python
import pandas as pd
from pandas import DataFrame
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('exam_score.xlsx')
print(data)
```

```python
x = DataFrame(data, columns = ['hours_studied'])
y = DataFrame(data, columns = ['exam_score'])
```

```python
regression = LinearRegression()
regression.fit(x,y)

regression.coef_ #theta_1

regression.intercept_ #
```

```python
plt.Figure(figsize = (10,6))
plt.scatter(x,y,alpha = 0.3)
plt.plot(x,regression.predict(x), color = 'red', linewidth = 2)
plt.title('hours_studied vs exam_score')
plt.xlabel('Hours_studied')
plt.ylabel('exam_score')
plt.xlim(1,12)
plt.ylim(50,90)
plt.show()
```
