```python
import matplotlib.pyplot as plt
```

```python
days = ['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
temperature = [50,47,30,35,52,48,45]
```

```python
%matplotlib inline

plt.figure(figsize=(10,4))
plt.plot(days, temperature, marker='o', color='#0abde3', linestyle='--', linewidth=2)
plt.title('Temperature over 7 days')
plt.xlabel("Days",fontsize =13 )
plt.ylabel("Temperature",fontsize =13)
plt.xlim(-0.1)
plt.style.use('classic')
plt.show()
```

```python
%matplotlib inline


categories = ['Food', 'Transport', 'Entertainment', 'Bills', 'Shopping']
expenses = [150, 50, 80, 120, 100]

plt.figure(figsize=(10, 5))
plt.bar(categories, expenses, color='#ff9f1a')
plt.title('Weekly Expenses ($)')
plt.xlabel('Category')
plt.ylabel('Amount ($)')
plt.style.use('fivethirtyeight')
plt.show()
```
