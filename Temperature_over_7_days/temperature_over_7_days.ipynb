```python
import matplotlib.pyplot as plt
```

```python
days = ['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
temperature = [50,47,30,35,52,48,45]
```

```python
%matplotlib inline

plt.figure(figsize=(10,5))
plt.plot(days, temperature, marker='o', color='#0abde3', linestyle='--', linewidth=2)
plt.title('temperature over 7 days')
plt.xlabel("Days",fontsize =13 )
plt.ylabel("Temperature",fontsize =13 )

plt.grid(True)
plt.style.use('classic')
plt.show()
```

```python
%matplotlib inline


plt.figure(figsize=(10,5))
plt.bar(days, temperature, color=['#ff9f1a'] )
plt.title('Temperature over 7 days', fontsize = 17)
plt.xlabel("Days",fontsize =13 )
plt.ylabel("Temperature",fontsize =13 )

plt.grid(True)
plt.style.use('classic')
plt.show()
```
