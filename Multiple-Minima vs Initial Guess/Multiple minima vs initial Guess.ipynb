# Example 2 : Multiple Minima vs Initial Guess & Advanced Functions 
## $$g(x) =x^4 - 4x^2 + 5$$ 

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# make some data
x1 = np.linspace(-2,2,1000)

def f(x):
    return x**4 - 4*x**2 + 5

def df(x):
    return 4*x**3 - 8*x
```

```python
#plot x1 and df(x) side by side
plt.figure(figsize=[15,5])

#Chart: Cost-Function
plt.subplot(1,2,1)

plt.title('Cost Function', fontsize = 17)
plt.xlabel('X', fontsize = 13)
plt.ylabel('f(X)',fontsize = 13)
plt.xlim(-2 ,2)
plt.ylim(0.5,5.5)
plt.plot(x1, f(x1), color = '#9c88ff', linewidth= 3)

#Chart: Derivative
plt.subplot(1,2,2)
plt.title('Slope of the Cost-Function', fontsize = 17)
plt.xlabel('X', fontsize = 13)
plt.ylabel('df(X)',fontsize = 13)

plt.xlim(-2,2)
plt.ylim(-6,8)

plt.plot(x1,df(x1), color= '#9c88ff', linewidth = 3)
plt.style.use('fivethirtyeight')
plt.show()
```

## Gradient-Descent as Python Function

```python
# Gradient Descent
def gradient_descent(derivative_func,initial_guess,multiplier = 0.02,precision = 0.001):
    new_x = initial_guess

    xlist = [new_x]
    slopelist = [derivative_func(new_x) ]
 
    for n in range(1500):
        previous_x = new_x
        gradient = df(previous_x)
        new_x = previous_x - multiplier * gradient

        xlist.append(new_x)
        slopelist.append(derivative_func(new_x))
        
        step_size = abs(new_x - previous_x)
        if step_size < precision:
            break
    
    return new_x, xlist, slopelist
```

```python
localminima , listx , drivlist = gradient_descent(derivative_func =df,initial_guess=0.1)
                                                  
print('Local Minima: ', localminima )
print('Number of step: ', len(listx ))
print('Slope : ', drivlist )
```

```python

localminima , listx , drivlist = gradient_descent(derivative_func =df,initial_guess=0.2)
#plot x1 and df(x) side by side
plt.figure(figsize=[15,5])

#Chart: Cost-Function
plt.subplot(1,2,1)

plt.title('Cost Function', fontsize = 17)
plt.xlabel('X', fontsize = 13)
plt.ylabel('f(X)',fontsize = 13)
plt.xlim(-2 ,2)
plt.ylim(0.5,5.5)
plt.plot(x1, f(x1), color = '#9c88ff', linewidth= 3)
plt.scatter(listx,f(np.array(listx)), color = 'red', s = 100, alpha = 0.5)
#Chart: Derivative
plt.subplot(1,2,2)
plt.title('Slope of the Cost-Function', fontsize = 17)
plt.xlabel('X', fontsize = 13)
plt.ylabel('df(X)',fontsize = 13)

plt.xlim(-2,2)
plt.ylim(-6,8)

plt.plot(x1,df(x1), color= '#9c88ff', linewidth = 3)
plt.scatter(listx , drivlist, color = 'red', alpha=0.5)
plt.style.use('fivethirtyeight')
plt.show()
```
