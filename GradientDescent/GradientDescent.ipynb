## Notebook modules and Packages

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Example - 1
$f(x) = x^2 + (x+1)$

```python
def f(x):
    return x**2 + (x + 1)
```

```python
#mark data
x1 = np.linspace(start = -3, stop = 3, num = 100)
```

```python
#make a graph
plt.figure(figsize=(7,5))
plt.xlabel('X', fontsize = 13)
plt.ylabel('f(X)',fontsize = 13)
plt.ylim(0,8)
plt.plot(x1, f(x1))
plt.style.use('fivethirtyeight')
plt.show()
```

```python
def df(x):
    return 2*x + 1
```

```python
#plot x1 and df(x) side by side
plt.figure(figsize=[15,5])

#Chart: Cost-Function
plt.subplot(1,2,1)

plt.title('Cost Function', fontsize = 17)
plt.xlabel('X', fontsize = 13)
plt.ylabel('f(X)',fontsize = 13)
plt.xlim(-3 ,3)
plt.ylim(0,8)
plt.plot(x1, f(x1), color = 'blue', linewidth= 3)

#Chart: Derivative
plt.subplot(1,2,2)
plt.title('Slope of the Cost-Function', fontsize = 17)
plt.xlabel('X', fontsize = 13)
plt.ylabel('df(X)',fontsize = 13)

plt.xlim(-2,3)
plt.ylim(-3,6)

plt.plot(x1,df(x1), color= 'blue', linewidth = 3)
plt.style.use('fivethirtyeight')
plt.show()
```

## Python loops and Gradient descent

```python
# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

xlist = [new_x]
slopelist = [df(new_x) ]
for n in range(1500):
    previous_x= new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient

    xlist.append(new_x)
    slopelist.append(df(new_x))
    
    step_size = abs(new_x - previous_x)
    if step_size < precision:
        print("loop run time: ",n)
        break

print('Local Minimum at: ',new_x)
print('Slope: ', df(new_x))
print('Cost_function: ',f(new_x))
```

```python
#super impose the Gradient Descent Calculations

plt.figure(figsize=[20,5])

#Chart: Cost-Function
plt.subplot(1,3,1)

plt.title('Cost Function', fontsize = 17)
plt.xlabel('X', fontsize = 13)
plt.ylabel('f(X)',fontsize = 13)
plt.xlim(-3 ,3)
plt.ylim(0,8)
plt.plot(x1, f(x1), color = 'blue', linewidth= 3, alpha = 0.8)

values = np.array(xlist)
plt.scatter(xlist, f(values),color ='red', s=100, alpha = 0.6)

#Chart: Derivative
plt.subplot(1,3,2)
plt.title('Slope of the Cost-Function', fontsize = 17)
plt.xlabel('X', fontsize = 13)
plt.ylabel('df(X)',fontsize = 13)

plt.xlim(-2,3)
plt.ylim(-3,6)

plt.plot(x1,df(x1), color= 'blue', linewidth = 3, alpha =0.8)

plt.scatter(xlist, slopelist, color = 'red', s = 100, alpha = 0.6)

#3: Gradient Descent
plt.subplot(1,3,3)
plt.title('Gradient Descent')
plt.xlabel('X', fontsize = 13)
plt.ylabel('df(X)',fontsize = 13)

plt.xlim(-0.55,3)
plt.ylim(-0.3,6)

plt.plot(x1,df(x1), color= 'blue', linewidth = 3, alpha =0.8)

plt.scatter(xlist, slopelist, color = 'red', s = 100, alpha = 0.6)
plt.style.use('fivethirtyeight')
plt.show()
```
